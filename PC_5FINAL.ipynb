## Ejercicio 4 

```python
# Reproyectar a EPSG:24891 (PSAD56 / Perú West Zone)
Peru_municipalities_projected = Peru_municipalities.to_crs(epsg=24891)

# Calcular el centroide de cada geometría en el CRS proyectado
Peru_municipalities_projected['centroid'] = Peru_municipalities_projected.geometry.centroid
```

```python
#Spatial Overlay 
# the north
MunisN_Peru=Peru_municipalities_projected.cx[:,centroidY:]
# the south
MunisS_Peru=Peru_municipalities_projected.cx[:,:centroidY]
# the west
MunisW_Peru=Peru_municipalities_projected.cx[:centroidX,:]
# the east
MunisE_Peru=Peru_municipalities_projected.cx[centroidX:,:]
```

```python
#Lets PLOT 
base=MunisN_Peru.plot(facecolor='green', edgecolor='black',linewidth=0.2, alpha=0.6)
MunisS_Peru.plot(facecolor='grey', edgecolor='black',linewidth=0.2,ax=base, alpha=0.4)
```

```python
base=MunisE_Peru.plot(facecolor='green', edgecolor='black',linewidth=0.2, alpha=0.6)
MunisW_Peru.plot(facecolor='grey', edgecolor='black',linewidth=0.2,ax=base, alpha=0.4)
```

```python
#Intersección 
```

```python
munisNS_Peru=MunisN_Peru.overlay(MunisS_Peru, how="intersection",keep_geom_type=True)
munisNS_Peru.plot()
```

```python
MunisN_Peru.sjoin(MunisS_Peru, how="inner", predicate='contains').plot()
```

```python
munisWE_Peru=MunisW_Peru.overlay(MunisE_Peru, how="intersection",keep_geom_type=True)
munisWE_Peru.plot(edgecolor='white',linewidth=0.1)
```
